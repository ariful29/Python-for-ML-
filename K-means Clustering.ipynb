```
#Dataset

import matplotlib.pyplot as plt
from random import random

x = []
y = []

for i in range(30):
  x.append(random()) 
  y.append(random()) 

for i in range(30):
  x.append(-random()) 
  y.append(-random()) 

for i in range(5):
  x.append(random()) 
  y.append(-random())

for i in range(5):
  x.append(-random()) 
  y.append(random())  

plt.scatter(x, y, color = "green")
plt.show()

c1x = random()
c1y = random()
c2x = -random()
c2y = -random()

plt.scatter(x, y, color = "green")
plt.scatter(c1x,c1y, color ="red")
plt.scatter(c2x,c2y, color ="blue")
plt.show()


#distance from centroid step-2
for i in range(10):
  d1=[]
  for k in range(70):
    d=(c1x-x[k])**2 + (c1y-y[k])**2
    d1.append(d)

  d2=[]
  for k in range(70):
    d=(c2x-x[k])**2 + (c2y-y[k])**2
    d2.append(d)

  # step-3
  count1=0
  count2=0
  sum1x=0
  sum1y=0
  sum2x=0
  sum2y=0

  for i in range(len(x)):
    if (d1[i]<d2[i]):
      plt.scatter(x[i],y[i],color="red")
      sum1x=sum1x+x[i]
      sum1y=sum1y+y[i]
      count1=count1+1

    else:
      plt.scatter(x[i],y[i],color="blue")
      sum2x=sum2x+x[i]
      sum2y=sum2y+y[i]
      count2=count2+1

  avg1x=sum1x/count1
  avg1y=sum1y/count1
  avg2x=sum2x/count2
  avg2y=sum2y/count2

  c1x = avg1x
  c1y = avg1y
  c2x = avg2x
  c2y = avg2y

  plt.scatter(c1x,c1y, color ="red", marker="x")
  plt.scatter(c2x,c2y, color ="blue",marker="x") 
  plt.show()



```
